```python

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python



url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key
```

```python
country = []
name = []
city = []
Date = []
latitude = []
longitude = []
max_temp = []
humidity = []
clouds = []
windy = []
 


```

```python
print("Beginning Data Retrieval")
print("-----------------------------")
```

```python

print("Beginning Data Retrieval")
print("-----------------------------")


Data_cities = []

record = 1 
set_count = 1

for city in cities:
    data_url = url + "&q=" + city
    print("Processing Record %s of Set %s | %s" % (record, set_count, city))

    
    try:
        weather_status = requests.get(data_url).json()
        City = weather_status["name"]
        country = weather_status["sys"]["country"]
        Date = weather_status["dt"]
        latitude = weather_status["coord"]["lat"]
        longitude = weather_status["coord"]["lon"]
        max_temp = weather_status["main"]["temp_max"]
        humidity = weather_status["main"]["humidity"]
        clouds = weather_status["clouds"]["all"]
        windy = weather_status["wind"]["speed"]
        
        
        
        
    except:
        print("Not Found")
        pass
              


print("Data Retrieval Complete      ")
print("-----------------------------")
        
```

```python
# Export the city data into a .csv.


# Display the DataFrame
weather_dict = {
    "Country":country,
    "City":name,
    "Date":Date,
    "Latitude":latitude,
    "Longitude":longitude,
    "Max Temp": max_temp,
    "Humidity":humidity,
    "Clouds":clouds,
    "Winds":windy
}
    
    
    
    

    
    
    

```

```python
weather_data = pd.DataFrame([weather_dict])
```

```python
weather_data.head()
```

```python
# Latitude vs. Temperature Plot


plt.scatter(weather_data["Latitude"], weather_data["Max Temp"], marker="o", s=10)

plt.title("City Latitude vs. Max Temperature")
plt.ylabel("Max. Temperature (F)")
plt.xlabel("Latitude")
plt.grid(True)

plt.show()
```

```python
plt.scatter(weather_data["Latitude"], weather_data["Humidity"], marker="o", s=8)
plt.title("City Latitude vs. Humidity")
plt.ylabel("Humidity (%)")
plt.xlabel("Latitude")
plt.grid(True)
plt.show()
```
